```python
import pandas as pd

# Basic read
df = pd.read_csv('data/SGJobData.csv')
```

```python
# View the first 5 rows
# Returns the company with the most job post entries
top_by_postings = df['postedCompany_name'].value_counts().idxmax()
count = df['postedCompany_name'].value_counts().max()

print(f"Top Employer (Most Posts): {top_by_postings} with {count} posts")

top_5 = df.groupby('postedCompany_name').agg({
    'metadata_jobPostId': 'count',
    'numberOfVacancies': 'sum',
    'average_salary': 'mean'
}).rename(columns={'metadata_jobPostId': 'Total_Posts'}).sort_values('Total_Posts', ascending=False)

print(top_5.head(5))
# Sum total applications per company and find the top one
popular_employer = df.groupby('postedCompany_name')['metadata_totalNumberJobApplication'].sum().idxmax()
total_apps = df.groupby('postedCompany_name')['metadata_totalNumberJobApplication'].sum().max()

print(f"Most Popular Employer: {popular_employer} ({total_apps} total applications)")
```

```python
popularity_rank = df.groupby('postedCompany_name').agg({
    'metadata_totalNumberJobApplication': 'sum',
    'metadata_totalNumberOfView': 'sum',
    'metadata_jobPostId': 'count'
}).sort_values('metadata_totalNumberJobApplication', ascending=False)

print(popularity_rank.head(5))
```

```python
# Sum total applications per category
popular_categories = df.groupby('categories')['metadata_totalNumberJobApplication'].sum().sort_values(ascending=False)

print(popular_categories.head(5))
```

```python
# 1. Split the strings into lists
df['categories_list'] = df['categories'].str.split(',')

# 2. "Explode" the lists into individual rows
df_exploded = df.explode('categories_list')

# 3. Now group by the individual categories
individual_cat_popularity = df_exploded.groupby('categories_list')['metadata_totalNumberJobApplication'].sum().sort_values(ascending=False)

print(individual_cat_popularity.head(10))
```

```python
category_stats = df.groupby('categories').agg({
    'metadata_totalNumberJobApplication': 'sum',  # Total interest
    'metadata_totalNumberOfView': 'sum',          # Total reach
    'metadata_jobPostId': 'count',                # Total supply of jobs
    'average_salary': 'mean'                      # Pay level
}).rename(columns={
    'metadata_totalNumberJobApplication': 'Total_Apps',
    'metadata_totalNumberOfView': 'Total_Views',
    'metadata_jobPostId': 'Job_Count'
})

# Sort by most applied categories
print(category_stats.sort_values('Total_Apps', ascending=False).head(10))
```
